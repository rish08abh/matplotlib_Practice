```python
import matplotlib.pyplot as plt
```

```python
#basic line plot
x=[2,3,5,6,7]
y=[4,6,7,8,9]
plt.plot(x,y,marker="o",c='b',label="relation of x and y",linewidth=4)
plt.title("First plot")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.legend()
plt.show()
```

```python
#scatter plot
x1=[2,3,4,5,6,7,8]
y1=[4,5,6,7,0.5,1,0]
plt.scatter(x1,y1,marker='o',c='black',label="scatter plot of x and y")
plt.title("Second Plot")
plt.legend()
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.grid(c="blue")
plt.show()
```

```python
#bar plot
x=[1,2,3,4]
y=[10,15,12,19]
plt.bar(x,y,label="bar graph",color='green')
plt.title("Third Plot")
plt.legend()
plt.show()
```
